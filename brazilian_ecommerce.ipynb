```python
df['order_period'] = df.order_purchase_timestamp.apply(lambda x: x.strftime('%Y-%m'))
df.head()
```

```python
df.set_index('customer_unique_id', inplace=True)
df['cohort_group'] = df.groupby(level = 0)['order_purchase_timestamp'].min().apply(lambda x: x.strftime('%Y-%m'))
df.reset_index(inplace=True)
df.head()
```

```python
grouped = df.groupby(['cohort_group', 'order_period'])
# count the unique users, orders, and total revenue per Group + Period
cohorts = grouped.agg({'customer_unique_id': pd.Series.nunique,
                       'order_id': pd.Series.nunique,
                       'payment_value': np.sum})
# make the column names more meaningful
cohorts.rename(columns={'customer_unique_id': 'total_users',
                        'order_id': 'total_orders'}, inplace=True)
cohorts.head()
```

```python
def cohort_period(df):
    df['cohort_period'] = np.arange(len(df)) + 1
    return df
cohorts = cohorts.groupby(level=0).apply(cohort_period)
cohorts.head()
```

```python
# reindex the DataFrame
cohorts.reset_index(inplace=True)
cohorts.set_index(['cohort_group', 'cohort_period'], inplace=True)

# create a Series holding the total size of each CohortGroup
cohort_group_size = cohorts['payment_value'].groupby(level=0).first()
cohort_group_size.head()
```

```python
cohorts['payment_value'].unstack(0).head()
```

```python
user_retention = cohorts['payment_value'].unstack(0).divide(cohort_group_size, axis=1)
user_retention.head(10)
```

```python
sns.set(style='white')
plt.figure(figsize=(14,6))
ax = plt.axes()
sns.heatmap(user_retention.T, mask=user_retention.T.isnull(), annot=True, fmt='.0%', ax = ax)
ax.set_ylim(22.0, 0);
```
